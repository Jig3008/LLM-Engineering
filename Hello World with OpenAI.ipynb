This is the first notebook where we explore on how to call the OpenAI agent from our code with OpenAI Key

```python
#All imports

import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
#Now we need to check if our dotenv file is working or not

load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("OPENAI_API_KEY is not set")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!") 
    
```

The OpenAI expects the prompt to be sent in the below format. A list of dictionaries.

```python
message = 'I want to groom my pet Shih Tzu in a professional way'

messages = [{
    'role' : 'user',
    'content' : message
}]

messages
```

```python
openai = OpenAI()
response = openai.chat.completions.create(model='gpt-4o-mini',messages=messages)
response.choices[0].message.content
```

For Sophisticated models, we use different types of Prompts
1. Role : User
    This is a conversation starter that they should reply to
2. Role: System
    This is informing the model what task they are performing and what tone they should use

```python
#Example

messages1 = [
    {
        'role' : 'system',
        'content' : 'You are a cheerful assistant that always uses emojis. Helps the user with their basic queries'
    },
    {
        'role' : 'user',
        'content' : 'I am feeling down today, can you cheer me up?'
    }
]

response1 = openai.chat.completions.create(model='gpt-4o-mini',messages=messages1)
response1.choices[0].message.content

```

```python
messages2 = [
    {
        'role' : 'system',
        'content' : 'You are a sarcastic assistant that answers questions with a lot of sarcasm'
    },
    {
        'role' : 'user',
        'content' : 'I am feeling down today, can you cheer me up?'
    }
]

response2 = openai.chat.completions.create(model='gpt-4o-mini',messages=messages2)
response2.choices[0].message.content
```

```python
#All Imports

import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown,display,update_display
import requests
```

```python
#Check if we have the openAI API key
load_dotenv(override=True)
openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key is None:
    print("No API Key found, please set the OPENAI_API_KEY environment variable")
elif not openai_api_key.startswith("sk-"):
    print("Invalid API Key, please check your OPENAI_API_KEY environment variable")
else:
    print("API Key found, proceeding...")
```

```python
ONLINE_MODEL = 'gpt-4o-mini'
OFFLINE_MODEL = 'phi3:latest'
```

```python
def create_prompts(system_prompt_style, user_prompt):
    SYSTEM_PROMPT = f"""
        You are a {system_prompt_style} type of assitant. You know most of the concepts of the world.
        The user would ask you a questions and you need to answer them with the best of your knowledge.
    """
    USER_PROMPT = f"""
        The user would ask you a question. You need to answer them with the best of your knowledge.
        The question is: {user_prompt}
    """
    return SYSTEM_PROMPT, USER_PROMPT
```

```python
def get_model_answer(user_question, model_name, needs_model_name, model_answer_style):
    if needs_model_name == True:
        requests.get("http://localhost:11434").content
        ollama_url = "http://localhost:11434/v1"
        openai = OpenAI(base_url= ollama_url, api_key="openaikey")
    else:
        openai = OpenAI()
    system_prompt, user_prompt = create_prompts(model_answer_style,user_question)
    stream_response = openai.chat.completions.create(model = model_name,
    messages=[{
        'role': 'system',
        'content' : system_prompt
    },{
        'role' : 'user',
        'content' : user_prompt
    }],
    stream=True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream_response:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response),display_id=display_handle.display_id)
```

Test 1:

Test 2:

- Style: Strict Teacher
- Model: gpt-oss:latest - using LLama

```python
def run_program():
    while True:
        MODEL = ""
        needs_model_name = False
        user_input = input('What is your question?: [Type done if you do not want to ask anything]')
        if user_input.lower() == 'done':
            break
        else:
            answer_style = input('What is your preferred answer style?:')
            model_type = input('Which model you want to use? [Offline/Online]:')
            if model_type.lower() == 'offline':
                MODEL = OFFLINE_MODEL
                needs_model_name = True
                get_model_answer(user_input,MODEL,needs_model_name,answer_style)
            elif model_type.lower() == 'online':
                MODEL = ONLINE_MODEL
                get_model_answer(user_input,MODEL,needs_model_name,answer_style)            
            else:
                print('Try again!')
                continue
```

```python
run_program()
```

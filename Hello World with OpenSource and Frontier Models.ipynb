Question 1: Why did we want to get the Open AI Package?

```python
#Import packages
import os
from dotenv import load_dotenv
import requests
#Get and check the Open AI API key

load_dotenv(override=True)
open_ai_key = os.getenv("OPENAI_API_KEY")

if not open_ai_key:
    print("No Open AI API key found")
elif not open_ai_key.startswith("sk-"):
    print("Open AI API key is not valid")
else:
    print("Open AI API key is valid")

#Without the Open AI package, we can still make the API call to Open AI

headers = {
    "Authorization": f"Bearer {open_ai_key}",
    "Content-Type": "application/json"
}

payload = {
    "model": "gpt-5-nano",
    "messages":[{"role": "user", "content": "Tell me a joke"}]
}
```

```python
#Now that we have the headers and payload, we can make the API call to Open AI

OPENAI_API_URL = "https://api.openai.com/v1/chat/completions"

response = requests.post(OPENAI_API_URL, headers=headers, json=payload)
response.json()
```

```python
response.json()["choices"][0]["message"]["content"]
```

Based on the above it is fair to conclude that we are able to do it but takes a longer route. And non-efficient at best. In comes the OpenAI package and the same code would take shorter and efficient readability

```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv(override=True)
open_ai_key = os.getenv("OPENAI_API_KEY")

if not open_ai_key:
    print("No Open AI API key found")   
elif not open_ai_key.startswith("sk-"):
    print("Open AI API key is not valid")
else:
    print("Open AI API key is valid")

messages = [{
    "role": "user",
    "content":"Tell me a joke"
}]

openai = OpenAI()
response = openai.chat.completions.create(model="gpt-5-nano",messages=messages)
response.choices[0].message.content


```

Question 2: Can we use the OPENAI package to work with other Model API

```python
load_dotenv(override=True)
google_api_key = os.getenv("GOOGLE_API_KEY")
GEMINI_API_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"

if not google_api_key:
    print("No Google API key found")
elif not google_api_key.startswith("AIza"):
    print("Google API key is not valid")
else:
    print("Google API key is valid")
```

```python
messages = [{
    "role": "user",
    "content":"Tell me a joke"
}]

googleapi = OpenAI(base_url=GEMINI_API_URL,api_key=google_api_key)

response = googleapi.chat.completions.create(model="gemini-2.5-flash-lite",messages=messages)
response.choices[0].message.content
```

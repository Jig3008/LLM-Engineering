```python
#All Imports

import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown,display,update_display
import requests
```

```python
load_dotenv(override=True)

def check_api_keys():
    openai_key = os.getenv("OPENAI_API_KEY")
    anthropic_key = os.getenv("ANTHROPIC_API_KEY")
    google_key = os.getenv("GOOGLE_API_KEY")
    grok_key = os.getenv("GROK_API_KEY")
    deepseek_key = os.getenv("DEEPSEEK_API_KEY")
    openrouter_key = os.getenv("OPENROUTER_API_KEY")
    
    if not openai_key:
        print("OpenAI API key is not set")
    if not anthropic_key:
        print("Anthropic API key is not set")
    if not google_key:
        print("Google API key is not set")
    if not grok_key:
        print("Grok API key is not set")
    if not deepseek_key:
        print("DeepSeek API key is not set")
    if not openrouter_key:
        print("OpenRouter API key is not set")
    
    print('If no issue, then all good to go')
    return openai_key, anthropic_key, google_key, grok_key, deepseek_key, openrouter_key
```

```python
def check_local_models():
    requests.get("http://localhost:11434").content
    print(requests.get("http://localhost:11434").content)
check_local_models()
```

```python
def set_all_base_urls_and_models():
    anthropic_url = "https://api.anthropic.com/v1/"
    gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
    deepseek_url = "https://api.deepseek.com"
    grok_url = "https://api.x.ai/v1"
    openrouter_url = "https://openrouter.ai/api/v1"
    ollama_url = "http://localhost:11434/v1"

    openai_key, anthropic_key, google_key, grok_key, deepseek_key, openrouter_key = check_api_keys()

    anthropic = OpenAI(base_url=anthropic_url, api_key=anthropic_key)
    google = OpenAI(base_url=gemini_url, api_key=google_key)
    deepseek = OpenAI(base_url=deepseek_url, api_key=deepseek_key)
    grok = OpenAI(base_url=grok_url, api_key=grok_key)
    openrouter = OpenAI(base_url=openrouter_url, api_key=openrouter_key)
    ollama = OpenAI(base_url=ollama_url, api_key='local api key')

    return anthropic, google, deepseek, grok, openrouter, ollama   
```

Test 1: Model Humor and Wittiness

```python
anthropic, google, deepseek, grok, openrouter, ollama = set_all_base_urls_and_models()

messages =[{
    'role': 'user',
    'content': 'Tell me a joke'
}]

anthropic_response = anthropic.chat.completions.create(model = 'claude-sonnet-4-6', messages = messages)
deepseek_response = deepseek.chat.completions.create(model = 'deepseek-chat', messages = messages)
grok_response = grok.chat.completions.create(model = 'grok-4-1-fast-non-reasoning', messages = messages)
ollama_response = ollama.chat.completions.create(model = 'gemma3:270m', messages = messages)

print('Anthropic: ', display(Markdown(anthropic_response.choices[0].message.content)))
print('DeepSeek: ', display(Markdown(deepseek_response.choices[0].message.content)))
print('Grok: ', display(Markdown(grok_response.choices[0].message.content)))
print('Ollama: ', display(Markdown(ollama_response.choices[0].message.content)))

```

```python
 #Now lets try with a more complex puzzle with dilemma

dilemma_prompt = """
You and a partner are contestants on a game show. You're each taken to separate rooms and given a choice:
Cooperate: Choose "Share" — if both of you choose this, you each win $1,000.
Defect: Choose "Steal" — if one steals and the other shares, the stealer gets $2,000 and the sharer gets nothing.
If both steal, you both get nothing.
Do you choose to Steal or Share? Pick one.
"""

dilemma = [
    {"role": "user", "content": dilemma_prompt},
]
```

```python
#We would be testing this with OpenAI, Anthropic, Grok and DeepSeek

openai_response = openai.chat.completions.create(model="gpt-4o-mini", messages=dilemma)
display(Markdown(openai_response.choices[0].message.content))
```

```python
anthropic_response = anthropic.chat.completions.create(model="claude-sonnet-4-6", messages=dilemma)
display(Markdown(anthropic_response.choices[0].message.content))
```

```python
grok_response = grok.chat.completions.create(model="grok-4-1-fast-non-reasoning", messages=dilemma)
display(Markdown(grok_response.choices[0].message.content))
```

```python
deepseek_response = deepseek.chat.completions.create(model="deepseek-chat", messages=dilemma)
display(Markdown(deepseek_response.choices[0].message.content))
```

OpenAI, Anthropic and DeepSeek - decided to Share
Grok took the aggressive route to Steal

```python
def check_humor_with_auto():
    print("Asking the Auto-Router for a joke...")
    
    response = openrouter.chat.completions.create(
        model="openrouter/auto", # The 'Magic' router
        messages=[{
            "role": "user", 
            "content": "Give me a deeply sarcastic joke about why AI will never replace middle management."
        }],
        extra_body={
            "route": "fallback"
        }
    )

    joke = response.choices[0].message.content
    
    actual_model = response.model 

    print(f"\n--- TEST RESULTS ---")
    print(f"MODEL USED: {actual_model}")
    print(f"HUMOR RESPONSE: \n{joke}")

check_humor_with_auto()
```

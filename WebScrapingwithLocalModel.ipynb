Day 2 exercise is about the scraping a webpage but with the help of a local machine based LLM. We would be using Ollama and model: gemini-3-flash-preview:latest

```python
#All Imports

import os
from openai import OpenAI
from IPython.display import Markdown,display
from scraper import fetch_website_contents
import requests
```

```python
requests.get('http://localhost:11434').content
```

```python
system_prompt = """
You are a web scraper who finds quirky and interesting information about a website. 
You are given the contents of a webpage and you
need to extract all meaningful information from the webpage. Feel free to remove
the header, navigational features of the page.
"""

user_prompt = """
Here are the contents of the webpage.
Can you find something quirky and interesting about the website?
"""
```

```python
def intialize_ollama():
    ollamaapi = OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")
    return ollamaapi
```

```python
def summarize_website(url,system_prompt,user_prompt):
    contents = fetch_website_contents(url)
    messages = [
        {"role":"system",
         "content":system_prompt},
        {"role":"user",
         "content":user_prompt + contents}
    ]
    ollama = intialize_ollama()
    response = ollama.chat.completions.create(model="gemini-3-flash-preview:latest",messages=messages)
    display(Markdown(response.choices[0].message.content))
```

```python
summarize_website('https://medium.com/@jigarjoshi7/why-your-data-is-probably-lying-to-you-and-how-not-to-kill-anyone-0e6091444d0c',system_prompt,user_prompt)
```
